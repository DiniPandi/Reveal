```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_real = pd.read_csv('Real.csv', encoding='cp1252')
data_disinfo = pd.read_csv('Disinfo.csv', encoding='cp1252')
```

```python
data_real.head()
```

```python
data_disinfo.head()
```

```python
data_real["class"] = 1
data_disinfo["class"] = 0
```

```python
data_real.shape, data_disinfo.shape
```

```python
data_real_manual_testing = data_real.tail(5)
for i in range(54,49,-1):
    data_real.drop([i], axis = 0, inplace = True)

data_disinfo_manual_testing = data_disinfo.tail(5)
for i in range(54,49,-1):
    data_disinfo.drop([i], axis = 0, inplace = True)
```

```python
data_real.shape, data_disinfo.shape
```

```python
data_real_manual_testing["class"] = 1
data_disinfo_manual_testing["class"] = 0
```

```python
data_real_manual_testing.head(5)
```

```python
data_disinfo_manual_testing.head(5)
```

```python
data_merge = pd.concat([data_real, data_disinfo], axis = 0)
data_merge.head(320)
```

```python
data_merge.isnull().sum()
```

```python
data_merge = data_merge.sample(frac = 1)
```

```python
data_merge.head()
```

```python
data_merge.reset_index(inplace = True)
data_merge.drop(['index'], axis = 1, inplace = True)
```

```python
data_merge.columns
```

```python
data_merge.head()
```

```python
def wordopt(Review):
    Review = Review.lower()
    return Review
```

```python
data_merge['Review'] = data_merge['Review'].apply(wordopt)
```

```python
x = data_merge['Review']
y = data_merge['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
xv_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state = 0)
GB.fit(xv_train, y_train)
```

```python
pred_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
# save the model
import pickle
filename = 'savedmodel.sav'
pickle.dump(RF, open(filename, 'wb'))
```

```python
load_model = pickle.load(open(filename,'rb'))
```

```python
def output_label(n):
    if n == 0:
        return "Review with Disinformation"
    elif n == 1:
        return "Genuine Review"

def manual_testing(review):
    testing_review = {"text":[review]}
    new_def_test = pd.DataFrame(testing_review)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)

    return print("\n\nLr Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}"
                 .format(output_label(pred_LR[0]),
                        output_label(pred_DT[0]),
                        output_label(pred_GB[0]),
                        output_label(pred_RF[0])))
```

```python
review = str(input())
manual_testing(review)
```
