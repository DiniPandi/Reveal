```python
df = df.drop(['User_id', 'Product_id', 'Rating', 'Date', 'Sentiment', 'Features'], axis=1)
```

```python
df
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
import re
```

```python
port_stem = PorterStemmer()
```

```python
port_stem
```

```python
def stemming(content):
    con = re.sub('[^a-zA-Z]',' ',content)
    con = con.lower()
    con = con.split()
    con = [port_stem.stem(word) for word in con if not word in stopwords.words('english')]
    con = ' '.join(con)
    return con
```

```python
df['Review'] = df['Review'].apply(stemming)
```

```python
x = df['Review']
y = df['Spam(1) and Not Spam(0)']
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
```

```python
vect = TfidfVectorizer()
```

```python
x_train = vect.fit_transform(x_train)
x_test = vect.transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

```python
for i, predict in enumerate(prediction):
    print(f"Prediction for sample {i+1}: {predict}")
```

```python
model.score(x_test, y_test)
```

```python
import pickle
```

```python
pickle.dump(vect, open('vector.pkl', 'wb'))
```

```python
pickle.dump(model, open('model.pkl', 'wb'))
```

```python
vector_form = pickle.load(open('vector.pkl', 'rb'))
```

```python
load_model = pickle.load(open('model.pkl', 'rb'))
```

```python
def detect_disinfo(review):
    review = stemming(review)
    input_data = [review]
    review_vector =  vector_form.transform(input_data)
    prediction = load_model.predict(review_vector)
    return prediction
```

```python
state = detect_disinfo("""The food at snack is a selection of popular Greek dishes. The appetizer tray is good as is the Greek salad. We were underwhelmed with the main courses.""")
```

```python
if state == [0]:
    print('Genuine Review')
else:
    print('This contains Disinformation')
```
